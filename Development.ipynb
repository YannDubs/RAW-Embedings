```python
%%cython
cimport cython
cimport numpy as np
import numpy as np

cpdef np.ndarray[double] exponential_rolling_average(avg,vec,alpha):
    return avg * (1-alpha) + vec * alpha

cpdef np.ndarray[double] weighted_average(new,old,weight):
    return new * weight + old * (1-weight)

cpdef raw_embed(str corpus, float weight=0.1, float alpha=0.3, int nEpoch=10, int dim=100):
    raw = {}
    rollingAverage = 0
    corpusSplit = corpus.split()
    cdef str w
    cdef int epoch

    for epoch in range(nEpoch):
        for w in corpusSplit:
            if w not in raw:
                raw[w] = np.random.uniform(-1,1,dim)
            new = weighted_average(rollingAverage,raw[w],weight) / (1 - weight)
            sign = np.sign(raw[w])
            raw[w] = np.minimum(np.abs(raw[w]),np.abs(new)) * sign
            rollingAverage = exponential_rolling_average(rollingAverage,raw[w],alpha)
        weight /= 1.1       
        
    return raw
```
